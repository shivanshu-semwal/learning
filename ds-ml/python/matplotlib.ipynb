# Matplotlib

### Importing the modules

```python
from matplotlib import pyplot as plt
```

### Creating a dataset

```python
x=['Rahul','Amit','Mudit']
y=[88,56,95]
```

### Plotting the Bar Graph

```python
plt.bar(x,y)
```

### Plotting the Scatter Graph

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.grid(True) #grid 
```

### New Dataset

```python
x=[16,8,10]
y=[8,16,6]
i=[8,15,11]
j=[6,15,7]
```

### Line graph

```python
plt.plot(x,y,'r',label='Line one',linewidth=5)
plt.plot(i,j,'m',label='Line two',linewidth=5)
plt.title('Information')
plt.legend() #give info.about line
plt.grid(True,color='g')
plt.show()
```

### New Dataset

```python
name=['Rohit','Shikhar','Virat','Hardik']
run=[50,87,80,65]
```

```python
plt.bar(name,run,color='green')
plt.title("Scour card")
plt.xlabel("Run")
plt.ylabel("Player")
plt.show()
```

### New Dataset

```python
x=[5,8,10]
y=[12,16,6]

i=[6,9,11]
j=[7,15,7]
```

```python
# different color of graph
plt.bar(x,y,color='y',align='center')
plt.bar(i,j,color='c',align='center')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```
