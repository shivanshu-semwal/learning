```python
class Video:
    def __init__(self, data):
        self.url = data['url']
        self.title = data['title']
        self.description = data['description']
        self.thumbnail = f"thumbnails/{data['thumbnail']}.webp"

    def _repr_html_(self):
        return f"""
        <div style="width: 200px">
            <a href="{self.url}">
                <img src="{self.thumbnail}"/>
            </a>
            {self.title}
        </div>
        """

    def __repr__(self):
        return str({
            "url": self.url,
            "title": self.title,
            "description": self.description,
            "thumbnail": self.thumbnail
        })
```

```python
class VideoList:
    def __init__(self, *videos):
        self.videos = videos

    def _repr_html_(self):
        return f"""
            <style>
                .container {{
                  display: flex;       
                  flex-direction: row; 
                  flex-wrap: wrap;     
                  gap: 10px;           
                  padding: 10px;
                }}
            </style>
            <div class="container">
                {' '.join(i._repr_html_() for i in self.videos)}
            </div>
        """
```

```python
videos_list = [Video(i) for i in data]
```

```python
import re
```

```python
categories = {
    "Coding Challenge": lambda video: bool(re.search("Coding Challenge", video.title, re.IGNORECASE)),
    "The Nature of Code": lambda video: bool(re.search(r"The Nature of Code|Nature of Code", video.title, re.IGNORECASE)),
    "p5.js Tutorial": lambda video: bool(re.search("p5.js Tutorial|p5.js", video.title, re.IGNORECASE)),
    "Processing Tutorial": lambda video: bool(re.search("Processing Tutorial", video.title, re.IGNORECASE)),
    "Programming with Text": lambda video: bool(re.search("Programming with Text", video.title, re.IGNORECASE)),
    "Intelligence and Learning": lambda video: bool(re.search(r"Intelligence and Learning|TensorFlow\.js Color Classifier", video.title, re.IGNORECASE)),
    "Topics of JavaScript/ES6": lambda video: bool(re.search(r"Topics of JavaScript/ES6|Topics of JavaScript/ES8", video.title, re.IGNORECASE)),
    "ml5.js": lambda video: bool(re.search("ml5.js", video.title, re.IGNORECASE)),
    "Git and GitHub for Poets": lambda video: bool(re.search("Git and GitHub for Poets", video.title, re.IGNORECASE)),
    "CircleCI": lambda video: bool(re.search("CircleCI", video.title, re.IGNORECASE)),
    "Working with Data and APIs in JavaScript": lambda video: bool(re.search("Working with Data and APIs in JavaScript", video.title, re.IGNORECASE)),
    "Guest Tutorial": lambda video: bool(re.search("Guest Tutorial", video.title, re.IGNORECASE)),
    "Q&A": lambda video: bool(re.search("Q&A", video.title, re.IGNORECASE)),
    "Live Stream": lambda video: bool(re.search("Live Stream", video.title, re.IGNORECASE)),
    "Twitter Bot Tutorial": lambda video: bool(re.search("Twitter Bot Tutorial", video.title, re.IGNORECASE)),
    "Discord Bots": lambda video: bool(re.search("Discord Bots", video.title, re.IGNORECASE)),
}
```

```python
categories_videos = {
    key: [i for i in videos_list if categories[key](i)] for key in categories.keys()
}
```
